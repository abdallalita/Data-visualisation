**Modify Line Style**

There are two ways of modifying the linestyle. 

The first uses a format string called **fmt** that has a colour, a line style and a marker as the values. So to change the marker, type and colour of a line, you would use, as an example, k--x. This makes the colour black, the line type a dashed line, and have x as a marker on the various data points. 

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, 'k--^') #modify line in the plot function

plt.show()
```

The second is a less confusing way of dealing with line types because it involves calling the characteristic directly by use of keyword arguements.

```python
plt.figure(figsize = (10, 5))#Default size is 6 by 4 inches

plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
plt.yticks(rotation = 60)

plt.ylabel('Loan Amount', fontsize = 15)  
plt.xlabel('Sector', fontsize = 15)

plt.plot(sector, loan, color = 'blue', linestyle = ':', marker = 'x')
plt.show()
```

**Plotting an additional variable on same plot**

Introduce another plot function

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Lender Count by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, c = '#EE66CC', ls = ':', marker = 'x')
plt.plot(sector, lender, c = '#22ff77')#add another plt.plot function

plt.show()
```

**Add a legend**


A better way is to include the `label` parameter in the `plt.plot()` method for each plot, then just call the `plt.legend()` method without any arguements. This way, it doesn't matter what order the graphs have been plotted, the legend will always point to the correct variable.

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Lender Count by Sector', fontsize = 15)

plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, c = '#EE66CC', ls = ':', marker = 'x', label = 'Loan Amount')
plt.plot(sector, lender, c = '#22ff77', label = 'Lender Count')#add another plt.plot function

plt.legend()

plt.show()
```

**Add a grid**

Use the `plt.grid()` method and pass in `True` as an arguement to generate a grid on your plot.

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Lender Count by Sector', fontsize = 15)

plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.plot(sector, loan, c = '#EE66CC', ls = ':', marker = 'x', label = 'Loan Amount')
plt.plot(sector, lender, c = '#22ff77', label = 'Lender Count')#add another plt.plot function

plt.legend()

plt.grid(True, color = 'k')

plt.show()
```

Sometimes a line plot might not do the work in presenting your data.Lets look at other plots in matplotlib

**Bar Plot**

Appropriate for plotting a categorical variable against a numerical variable.  


```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.bar(sector, loan, color = 'black')

plt.show()
```

By default the width of the bars is 0.8 inch. You can change that using the `width` parameter. Use `edgecolor` to define a boundary between the bars in the event they touch each other and you need to tell them apart.

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.bar(sector, loan, width = 0.2, color = '#b62708', edgecolor = 'blue')

plt.show()
```

**Combine bar plot and line plot**


```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Funded Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)
# We have added a legend using label and plt.legend.   
plt.plot(sector, fund, color = 'k', label = 'Funded Amount')#line plot
plt.bar(sector, loan, label = 'Loan Amount')#bar plot

plt.legend()

plt.show()
```

**Plot an additional variable as a bar plot**


```python
#Show them side by side 

1. We want to reduce the width of the bars so that we can be able to fit them side by side on the x axis.
2. Then shift the position of one bar by the width to be able to display the two bars side by side.

We generate an array of x labels using the np.arange(len()) method. This turns the categorical values on the x varaible(sector column) into an array of numbers which we can use to position the minibars on the x axis.

By default the width of bars is 0.8. We set the width  of the 2 Y variables to a smaller size (0.3), then again shift one of the Y variables by that width, leaving the other in place. This gives us the result we want.
```

```python
x_indices = np.arange(len(sector))
print(x_indices)
```

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount and Funded Amount by Sector', fontsize = 15)
plt.xlabel('Sector', fontsize = 15)
plt.ylabel('Loan Amount', fontsize = 15)

# converting sector variable to get the position of ticks to insert on the x axis, 
#This array is going to be used in place of the categorical variable sector to set the position.

x_indices = np.arange(len(sector))
width = 0.3

plt.bar(x_indices, loan, width = width, label = 'Loan Amount')
plt.bar((x_indices + width), fund, width = width, label = 'Funded Amount')

#Reset the xtick labels to the sector categories
plt.xticks(ticks = x_indices, labels = sector, rotation = 75)

plt.legend()

plt.show()
```

**Horizontal Bar Plot**


```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector', fontsize = 15)
plt.ylabel('Sector', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

plt.barh(sector, loan) 

plt.show()
```

To sort it in descending order, reverse the orders of the variables. 

We have to convert the columns(Which are of type Series) into lists, then reversed those lists, then make the plot using the reverse lists as the variables.

```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount by Sector', fontsize = 15)
plt.ylabel('Sector', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)

plt.xticks(rotation = 75)

#convert to list
sector_l = list(sector)
loan_l = list(loan)   

#Reverse lists
sector_l.reverse()
loan_l.reverse()

plt.barh(sector_l, loan_l)

plt.show()
```

**Scatter Plot**

Appropriate to show relationship between two numerical variables.


```python
plt.figure(figsize = (10,5))

plt.title('Loan Amount vs Lender Count', fontsize = 15)
plt.xlabel('Loan Amount', fontsize = 15)
plt.ylabel('Lender Count', fontsize = 15)

plt.xticks(rotation = 75)

plt.scatter(loan, lender)

plt.show()
```

**Histogram**

Appropriate to show the distribution of a variable.

By default it returns a plot with 10 bins.

```python
plt.figure(figsize = (10,5))

plt.title('Distribution of count', fontsize = 15)

plt.hist(df['count'], edgecolor = 'k', bins = 5)

plt.show()
```

**Multiple Plots**

Instead of having lots of information on one plot, we can choose to have multiple plots on one canvas, each showing one bit of information.

```python
# create a canvas and specify its size.
plt.figure(figsize = (15,5))

#1 row 2 columns and postion of the subplot.
plt.subplot(1,2,1)

plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
plt.xlabel('Sector')
plt.ylabel('Loan Amount')
plt.plot(sector, loan)

#1 row 2 columns and postion of the subplot.
plt.subplot(1,2,2) 
plt.title('Lender Count by Sector')
plt.xticks(rotation = 75)
plt.xlabel('Sector')
plt.ylabel('Lender Count')
plt.plot(sector, lender)
#save the plot canvas
plt.savefig('plot.png')
plt.show()
```
