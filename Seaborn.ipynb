seaborn is a visualization library that is better or more advanced than matplotlib,it makes visualization aesthetically more pleasing while adding some extra probabilities too. 
seaborn uses fewer code and has a stunning default themes.


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
#importing the dataset
ram=pd.read_csv("seaborn.csv")
```

```python
#line plot normal matplotlib
#Before setting seabornâ€™s theme
plt.plot(ram['loan_amount'], ram['lender_count'], color = 'r')
plt.show()
```

```python
#set the themes of the seaborn,grids
sns.set()
```

```python
plt.plot(ram['loan_amount'], ram['lender_count'], color = 'r')
plt.show()
```

```python
#Seaborn supports various themes that can make styling the plots really easy and save a lot of time.
#Using the set_style() function of Seaborn we can set any of the themes available on Seaborn library.
Here are a few of the popular themes:
â€¢ darkgrid
â€¢ whitegrid
â€¢ dark
â€¢ white
â€¢ ticks


#change the grids
sns.set_style("whitegrid")
```

```python
plt.plot(ram['loan_amount'], ram['lender_count'], color = 'r')
plt.show()
```

**Barplots**


```python
#using seaborn functions to plot with error bars (black lines) in certain confidence interval for a sample.
sns.barplot(x = 'sector', y = 'loan_amount', data = ram)
plt.show()
```

```python
#remove the lines(error bars) cause here it is a population
sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci=None)
plt.show()
```

```python
#using matplotlib functions to customize our plot

#We are going to use Matplotlibâ€™s functions to customize our plot.
plt.figure(figsize = ) to resize the plot
plt.title() to add a title
plt.xticks() method to rotate the labels on the x axis


plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci = None)
plt.show()
```

```python
#Customize plot
plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)

#Add labels to the bar and plot
ax = sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci = None)
for p in ax.patches:
    ax.annotate("%.2f" % p.get_height(), (p.get_x() + p.get_width() / 2., p.get_height()),                 
                ha = 'center', va = 'center', rotation = 60, xytext = (0, 20), textcoords = 'offset points')
plt.show()
```

```python
#Customization on bars is however done inthe plotting function.
#Specify the colour of the bars with the color parameter

plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci = None, color ='lightblue')
plt.show()
```

```python
#Change estimator.
#By default seaborn barplot shows the mean of the values and not the totals.

#You can have the graph show the totals (np.sum), median (np.median), etc by the altering the
#estimator parameter.

plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci = None, palette ='Dark2',estimator = np.sum)
plt.show()
```

```python
plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
sns.barplot(y = 'sector', x = 'loan_amount', data = ram, ci = None, color ='lightblue',
estimator = np.std)
plt.show()
```

```python
# use the hue parameter to add an extra dimension to the plot by coloring the bars accordingto a third variable.
#In barchart this will automatically split the bars into the distinct number of items based on that variable.
plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci = None,estimator = np.sum, hue = 'gender_lead')
plt.show()
```

```python
#Using a different column, this time repayment interval, shows that the bars will be split t0
#the number of unique categories in that column, in this case, 4. The colour is also automatically assigned

plt.figure(figsize = (10,5))
plt.title('Loan Amount by Sector')
plt.xticks(rotation = 75)
sns.barplot(x = 'sector', y = 'loan_amount', data = ram, ci = None,
estimator = np.sum, hue = 'repayment_interval')
plt.show()
```

**Relational Plot**

```python
plt.figure(figsize = (10,5))
sns.scatterplot(x = 'loan_amount', y = 'lender_count', data = ram)
plt.show()
```

```python
#We can also include hue parameter to add another dimension to our plot
#Pass in sector to hue to get markers(data points) coloured depending on sector

plt.figure(figsize = (15,10))
sns.scatterplot(x = 'loan_amount', y = 'lender_count', data = ram, hue ='sector')
plt.show()
```

```python
#We can have rows arranged columnwise by using col parameter
#We can use the main parent plotting function relplot() together with col and row parameters
#to separate plots by categorical classes. (instead of overloading semantics on one plot)

sns.relplot(x = 'loan_amount', y = 'lender_count', data = ram, col = 'repayment_interval', aspect = 1)
plt.show()
```

```python
# We can have rows arranged rowise by using row parameter
sns.relplot(x = 'loan_amount', y = 'lender_count', data = ram,row = 'repayment_interval', aspect = 1.5)
plt.show()
```

```python
#use both row and col to include two extra variables.One will be plotted columnwise and the other row-wise
sns.relplot(x = 'loan_amount', y = 'lender_count', data = ram,
row = 'gender_lead', col = 'repayment_interval', aspect = 1.5, hue ='sector', size = 'loan_amount', sizes = (100,500))
plt.show()
```

**Distribution Plots**

distplot(),histplot()

distplot(),which is the parent function and by default results in a histogram fitted with a kernel density
estimate (In statistics, kernel density estimation (KDE) is a non-parametric way to estimate the probability density function of a random variable. Kernel densityestimation is a fundamental data smoothing problem where inferences about the population are made, based on a finite data sample

```python
sns.distplot(ram['average_count'])
plt.show()
```

```python
#Histogram
sns.distplot(ram['average_count'], kde=False)
plt.show()
```

```python
#We can set bins to 20

sns.distplot(ram['average_count'], kde = False, bins = 20)
plt.show()
```

```python
#Using histogram
sns.histplot(ram['average_count'], bins = 20)
plt.show()
```

**Pair Plot**

```python
#Visualizes the relationships between every numerical variable in the data frame.
#Each column is compared to every other column (including itself to see what interesting relationships may exist between combinations of variables 
sns.pairplot(ram)
plt.show()
```
